```python
from sklearn.datasets import load_iris
from IPython.display import display
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
iris_data = load_iris(as_frame=True)
iris_dataframe = iris_data['data']
display(iris_dataframe.head())
```

```python
iris_dataframe.shape
```

```python
target = iris_data['target']
target
```

```python
model = DecisionTreeClassifier(max_depth=None)
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_dataframe, target, test_size=0.3, random_state=42)
```

```python
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
cm = confusion_matrix(y_test, y_pred)
cm_dataframe = pd.DataFrame(cm, columns=['Iris-Setosa', 'Iris-Versicolour', 'Iris-Virginica'], index=['Iris-Setosa', 'Iris-Versicolour', 'Iris-Virginica'])
sns.heatmap(cm_dataframe, annot=True)
```
